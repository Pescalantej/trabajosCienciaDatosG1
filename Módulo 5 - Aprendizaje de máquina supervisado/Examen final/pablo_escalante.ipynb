```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
datos_vinos = load_wine()
datos_vinos
print(datos_vinos['DESCR'])
```

```python
datos_vinos_DF = pd.DataFrame(datos_vinos.data)
datos_vinos_DF['objetivo'] = datos_vinos.target
datos_vinos_DF
```

### Desarrollo
Consideraciones preliminares:
- Se trabajará con un algoritmo *SVM (Support Vector Machine)* para crear un modelo de clasificación usando una función *kernel* de tipo lineal.

- Se empleara una normalización *MinMaxScaler* sobre las *features*.

- Para evaluar el clasificador, se utilizará validación cruzada utilizando como métrica *f1 score*. Además se evaluará si distintos valores del hiperparámetro *C (factor de tolerancia)* afectan el desempeño del modelo de clasificación.

- A modo de referencia visual, se creará una representación gráfica de dos *features* del dataset y las respectivas regiones definidas por los hiperplanos del modelo.

```python
# Cargamos librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from mlxtend.plotting import plot_decision_regions
```

```python
# Normalización de los datos
df_norm = datos_vinos_DF.copy()
df_norm.iloc[:,:-1] = preprocessing.MinMaxScaler().fit_transform(df_norm.iloc[:,:-1])
df_norm.sample(5)
```

```python
# Variables independientes
variables_indep = df_norm.drop('objetivo', axis = 1).columns
variables_indep
```

```python
# Datos X e y
df_norm_X = df_norm[variables_indep]
df_norm_y = df_norm['objetivo']
```

```python
# Conjuntos de train y test
df_norm_X_train, df_norm_X_test, df_norm_y_train, df_norm_y_test = train_test_split(
    df_norm_X, df_norm_y, test_size=0.3, random_state=42)
```

```python
# Creamos una instancia del modelo
clasificador = SVC(kernel= 'linear', C= 1.2)
```

```python
# Entrenamiento y cálculo de predicciones del modelo
predicciones = clasificador.fit(df_norm_X_train, df_norm_y_train).predict(df_norm_X_test)
```

```python
# Revisión de algunas predicciones del modelo
pd.concat([pd.DataFrame({'y_test':df_norm_y_test.reset_index(drop=True)}),
           pd.DataFrame(predicciones, columns=['y_pred'])], axis = 1).sample(10)
```

```python
# Vectores de soporte del modelo
print('Hay {} vectores de soporte.\nEjemplo de algunos vectores de soporte:'.format(len(clasificador.support_vectors_)))
pd.DataFrame(clasificador.support_vectors_).sample(5)
```

### Métrica de evaluación

```python
# Validación cruzada usando F1 score.
scores = cross_val_score(clasificador, df_norm_X, df_norm_y, cv=10, scoring='f1_micro')
print('Promedio F1: {:.3}\nDesviación estándar F1: {:.2}'.format(scores.mean(), scores.std()))
```

```python
# Influencia del valor de C en F1 score
C_values = np.geomspace(0.01, 30, num=100)
f1_values = []
for c in C_values:
    clasificador = SVC(kernel= 'linear', C= c)
    f1 = cross_val_score(clasificador, df_norm_X, df_norm_y, cv=10, scoring='f1_micro').mean()
    f1_values.append(f1)
```

```python
# Gráfico C vs F1
plt.figure(figsize=(8, 5))
plt.plot(C_values, f1_values)
plt.xlabel('C')
plt.ylabel('F1 score')
plt.grid()
plt.show()
```

```python
# Valores de C a los cuales F1 se hace máximo.
# Para valores de C = 1.2, 0.8 se consigue un F1 máximo.
c_f1_values = pd.DataFrame(list(zip(C_values, f1_values)), columns=['C', 'F1_score'])
c_f1_values.sort_values(by=['F1_score'], ascending = False).head(5)
```

### Representación gráfica

```python
# Se eligen dos features para graficar las regiones de decisión.
x_feature, y_feature = 6, 12
X_plot = np.array(df_norm[[x_feature,y_feature]])
y_plot = datos_vinos.target
clasificador.fit(X_plot,y_plot)

# Gráfico de regiones de decisión
plt.figure(figsize=(8, 5))
plot_decision_regions(X_plot,y_plot,clf=clasificador)
plt.xlabel(datos_vinos.feature_names[x_feature])
plt.ylabel(datos_vinos.feature_names[y_feature])
plt.xlim(-0.2, 1.2)
plt.ylim(-0.5, 1.5)
plt.show()
```

### Conclusiones
- El uso de la normalización *MinMaxScaler* proporcionó dos mejoras al momento de evaluar el modelo; Por un lado se consiguió un mayor *F1 score* y se redujeron considerablemente los tiempos de computo de los scritps.
- Durante el desarrollo de este trabajo se hicieron pruebas con las funciones *kernel* de base radial y polinomial, pero estas presentaron menor desempeño.
- El máximo valor de *F1 score* fue de 0.989, el cuál se consiguió para valores de C en torno a 1.2 y 0.8. Dichos resultados fueron obtenidos usando validación cruzada (*cross validation*) iterando sobre el factor de tolerancia C.
- El gráfico de las regiones de validación debe ser visto sólo como referencia, ya que en él solo se expone la influencia de dos *features* del dataset.
